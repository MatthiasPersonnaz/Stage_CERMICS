```julia
using LinearAlgebra;
using SparseArrays;
BLAS.get_config()
```

```julia
function laplacien(N)
    N² = N*N;
    # BLOC EXTRA-DIAGONAL
    J = -1.0*I(N)
    
    # BLOC DIAGONAL
    diag = +4*ones(Float64,N);  # vecteur de la diagonale
    extra_diag = -1.0*ones(Float64,N-1); # vecteur de la sur- et sous-diagonale remplies de 1
    T = SymTridiagonal(diag, extra_diag);  # matrice des blocs diagonaux, qui est tridiagonale symétrique

    # assemblage du laplacien
    Λ = spzeros(N²,N²);
    @views for i in 1:N
        Λ[1+(i-1)*N:i*N,1+(i-1)*N:i*N] .= T[:,:]
    end
    
    @views for i in 1:N-1
        Λ[1+(i-1)* N : i    *N, 1+(i)  *N :  (i+1)*N] .= J[:,:] # remplissage blocs extradiag supérieurs
        Λ[1+(i)  * N : (i+1)*N, 1+(i-1)*N :      i*N] .= J[:,:] # remplissage blocs extradiag inférieurs
    end
    h = 1/(N+1);
    return Λ # de taille N*N
end



#### séparation des deux contributions du laplacien 2D ####
function laplacian_2D_rescaled_dim_elec(N, N²) # partie du laplacien 2D pour la dimension électronique: manque juste le facteur 1/(m*δr²)
    Λ          = spzeros(N²,N²);
    diag       = 2*ones(Float64,N);        # termes uniquement sur la diagonale du bloc diagonal coefficientés de 2
    extra_diag = -ones(Float64,N-1);       # ainsi que sur l'extra diagonale du bloc diagonal coefficientés de -1
    T          = SymTridiagonal(diag, extra_diag);
    @views for i in 1:N
        Λ[1+(i-1)*N:i*N,1+(i-1)*N:i*N] .= T[:,:]
    end
    return Λ # symtridiagonal
end

function laplacian_2D_rescaled_dim_nucl(N, N²) # partie du laplacien 2D pour la dimension nucléaire: manque juste le facteur K*ϵ²/δu²
    Λ = spzeros(N²,N²)
    T = 2*sparse(I,N,N);     # termes uniquement sur la diagonale du bloc diagonal coefficientés de 2
    J = -sparse(I,N,N);      # ainsi que sur la diagonale du du bloc extra-diagonal coefficientés de -1
    @views for i in 1:N
        Λ[1+(i-1)*N:i*N,1+(i-1)*N:i*N] .= T[:,:]
    end
    @views for i in 1:N-1
        Λ[1+(i-1)* N : i    *N, 1+(i)  *N :  (i+1)*N] .= J[:,:]
        Λ[1+(i)  * N : (i+1)*N, 1+(i-1)*N :      i*N] .= J[:,:]
    end
    return Λ # sparse
end
```

```julia
n = 120; # nombre de points sur la grille
N = n*n; # dimension du vecteur

## VERSION MATRICE DE MON CAS SPÉCIFIQUE ##
β=1.5; η=.5; V0=1.5; σ=1.;
function V_nucl_nucl(R)
    return β/sqrt(η^2+R^2) 
end

function V_nucl_el(r,R)
    return  -V0*exp(-(r-R/2)^2/2/σ^2) + -V0*exp(-(r+R/2)^2/2/σ^2)
end


Λ2D_elec = laplacian_2D_rescaled_dim_elec(n,N);
Λ2D_nucl = laplacian_2D_rescaled_dim_nucl(n,N);
R₀ = 1.6267549186290133;
M = 500;
m = 1;
K = 0.391238712024284;
ϵ = 1/sqrt(sqrt(K*M));
ϵ² = ϵ*ϵ;
u_min = -2.65;
u_max = +2.65;    
r_min=-5.;
r_max=+5.;
δu = (u_max-u_min)/(n-1);
δr = (r_max-r_min)/(n-1);
δu² = δu*δu;
δr² = δr*δr;
us = Vector(u_min:δu:u_max);
rs = Vector(r_min:δr:r_max); 
ug = us' .* ones(n); 
rg = ones(n)' .* rs;



V_res = V_nucl_el.(rg, R₀.+ϵ*ug) + V_nucl_nucl.(R₀.+ϵ*ug); # (r,u) ↦ V(r,u) 

Λ  = K*ϵ²/2/δu²*Λ2D_nucl + 1/(2*m*δr²)*Λ2D_elec + Diagonal(reshape(V_res, N))


#### CONTRÔLE
@show typeof(Λ);
@show norm(convert(SparseMatrixCSC{Float32, Int64}, Λ)-Λ) 
@show ishermitian(Λ), isposdef(Λ);
```
