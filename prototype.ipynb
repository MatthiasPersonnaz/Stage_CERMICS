```julia
diff1_ctr_o8 =  [1/280 ‚àí4/105 1/5 ‚àí4/5 0. 4/5 ‚àí1/5 4/105 ‚àí1/280];
diff2_ctr_o8 = [‚àí1/560 8/315 ‚àí1/5 8/5 ‚àí205/72 8/5 ‚àí1/5 8/315 ‚àí1/560];
```

```julia
function assembly(T,J,N,N¬≤)
    Œõ = spzeros(N¬≤,N¬≤);
    @views for i in 1:N
        Œõ[1+(i-1)*N:i*N,1+(i-1)*N:i*N] .= T[:,:]
    end
    
    @views for i in 1:N-1
        Œõ[1+(i-1)* N : i    *N, 1+(i)  *N :  (i+1)*N] .= J[:,:] # remplissage blocs etradiag sup√©rieurs
        Œõ[1+(i)  * N : (i+1)*N, 1+(i-1)*N :      i*N] .= J[:,:] # remplissage blocs extradiag inf√©rieurs
    end
    return Œõ
end

function hamiltonian_2D(Œ¥r, Œ¥r¬≤, Œ¥R, Œ¥R¬≤, N, N¬≤, V, m, M)
    # BLOC EXTRA-DIAGONAL 
    J = -1/(2*M*Œ¥R¬≤)*sparse(I,N,N);
    
    # BLOC DIAGONAL
    diag = -(1/(m*Œ¥r¬≤)+1/(M*Œ¥R¬≤))*ones(Float64,N);  # vecteur de la diagonale RQ: le *2 sur la diagonale du laplacien discret se simplifie avec 1/2m
    extra_diag = 1/(2*m*Œ¥r¬≤)*ones(Float64,N-1);     # vecteur de la sur- et sous-diagonale remplies de 1
    T = - SymTridiagonal(diag, extra_diag);    # matrice des blocs diagonaux, qui est tridiagonale sym√©trique

    # assemblage du laplacien
    Œõ = assembly(T,J,N,N¬≤)
    V = Diagonal(reshape(V,N¬≤)); # a priori inutile de convertir en sparse avant

    H = Œõ+V;
    H, Œõ, V # op√©rateurs (matrices de taille N¬≤√óN¬≤)
end


function hamiltonian_1D(Œîd¬≤, N, V, m)
    Œõ = -1/(Œîd¬≤*2*m)*SymTridiagonal(-2*ones(Float64,N),ones(Float64,N-1));
    V = sparse(Diagonal(V)); # V_fun.(LinRange(r_min,r_max,N)) pour cr√©er l'argument
    H = Œõ + V; 
    H, Œõ, V
end


                                 Œ¥u, Œ¥u¬≤, Œ¥r, Œ¥r¬≤, N, N¬≤, V‚ÇÄ_res, m, œµ^2, K
function hamiltonian_2D_rescaled(Œ¥u, Œ¥u¬≤, Œ¥r, Œ¥r¬≤, N, N¬≤, V, m, œµ¬≤, K)
    # BLOC EXTRA-DIAGONAL 
    J = -K*œµ¬≤*1/(2*Œ¥u¬≤)*sparse(I,N,N); # /œµ¬≤ enlev√©
    
    # BLOC DIAGONAL
    diag = -(1/(m*Œ¥r¬≤)+K*œµ¬≤/Œ¥u¬≤)*ones(Float64,N); # vecteur de la diagonale RQ: le *2 sur la diagonale du laplacien discret se simplifie avec 1/2m
    extra_diag = 1/(2*m*Œ¥r¬≤)*ones(Float64,N-1);     # vecteur de la sur- et sous-diagonale remplies de 1
    T = - SymTridiagonal(diag, extra_diag);    # matrice des blocs diagonaux, qui est tridiagonale sym√©trique

    # assemblage du laplacien
    Œõ = assembly(T,J,N,N¬≤);
    V = Diagonal(reshape(V,N¬≤)); # a priori inutile de convertir en sparse avant

    H = Œõ + V;
    H, Œõ, V # op√©rateurs (matrices de taille N¬≤√óN¬≤)
end
```

On part d'un hamiltonien repr√©sentant par exemple, l'√©nergie de la mol√©cule H_2^+.
\begin{equation}
 -\frac{1}{2m} \frac{\partial^2}{\partial r^2} - \frac{1}{2M} \frac{\partial^2}{\partial R^2} + V(r,R)
\end{equation}
Qu'on d√©compose en trois termes
\begin{equation}
H = \hat{H}_x + \hat{H}_y + \hat{W}
\end{equation}

avec

\begin{equation}
\hat{H}^0_R =  - \frac{1}{2m} \frac{\partial^2}{\partial R^2} + \frac{1}{2} \frac{\mathrm{d}^2 E_0}{\mathrm{d}R^2}(R_0)(R-R_0)^2 
\end{equation}

\begin{equation}
        \hat{H}^0_r = - \frac{1}{2M} \frac{\partial^2}{\partial x^2} + V(r,R_0)   
\end{equation}

et la perturbation
\begin{equation}
\hat{W} = \frac{\partial V}{\partial R}(r,R_0)(R-R_0)  + \frac{1}{2} \left(\frac{\partial^2V}{\partial R^2}(r,R_0) -\frac{\mathrm{d}^2 E_0}{\mathrm{d}R^2}(R_0) \right)   (R-R_0)^2 + \text{h.o.t.}
\end{equation}
sur laquelle on n√©glige les termes de plus haut ordres (dans l'√©criture ci-dessus, on a n√©glig√© les termes d'ordre sup√©rieur ou √©gal √† 3)

```julia
me = 1; mp = 250;
M=(2*mp^3+mp^2*me)/(2*mp*(me+mp));
m=(2*mp^2*me)/(mp*(2*mp+me)); 
r_min=-5.; r_max=5.; R_min=0.0; R_max=3.5; N=150; œâ=1.; kdim=30; # augmenter la dimension Krylov quand N est grand: pour N=200, prendre kdim=50
Œ≤=1.5; Œ∑=.5; V0=1.5; œÉ=1.;


function V_nucl_el(r,R)
     return -V0*( exp(-(r-R/2)^2/2/œÉ^2) + exp(-(r+R/2)^2/2/œÉ^2) ) # potentiels interaction √©lectron avec les 2 noyaux
end

function V_nucl_nucl(R)
     return + Œ≤/sqrt(Œ∑^2+R^2) # potentiel interaction des 2 noyaux entre eux
end





Œ¥r = (r_max-r_min)/(N-1);
Œ¥R = (R_max-R_min)/(N-1);
Œ¥r¬≤ = Œ¥r*Œ¥r;
Œ¥R¬≤ = Œ¥R*Œ¥R;
N¬≤ = N*N;


# CONSTRUCTION DE LA MESHGRID
rs = Vector(r_min:Œ¥r:r_max); # direction coordonn√©e √©lectron
Rs = Vector(R_min:Œ¥R:R_max); # direction coordonn√©e distance noyau-noyau
rg = ones(N)' .* rs;
Rg = Rs' .* ones(N);
V = zeros(N,N);



# CONSTRUCTION DU POTENTIEL ORIGINAL ET DU HAMILTONIEN SUR GRILLE param√©tr√© en R
V = @. V_nucl_el(rg, Rg) + V_nucl_nucl(Rg); # potentiel d'interaction sur la grille tous ph√©nom√®nes compris
HÃÇ, Œõ, VÃÇ = hamiltonian_2D(Œ¥r, Œ¥r¬≤, Œ¥R, Œ¥R¬≤, N, N¬≤, V, m, M);
lowest_energies_y = zeros(N);



# STRUCTURE DU LAPLACIEN 1D
LS = SymTridiagonal(-2*ones(Float64,N),ones(Float64,N-1)); 

# LAPLACIENS SUR AXES param√©tr√©s en R
Œõr = -1/(Œ¥r¬≤*2*m)*LS;  # laplacien sur l'axe r
ŒõR = -1/(Œ¥R¬≤*2*M)*LS;  # laplacien sur l'axe R


# RECHERCHE DU R‚ÇÄ  minimisant l'√©nergie de l'√©tat fondamental
lE‚ÇÄ = zeros(N);
Base.Threads.@threads for j in 1:N
    Vx = sparse(Diagonal(Vector(V_nucl_el.(rs,R_min+j*Œ¥R)))) # potentiel en x √† R=jŒ¥R fix√© (i.e. √† distance noyau-noyau fix√©)
    vals, ~, infos = KrylovKit.eigsolve(Œõr+Vx, N, 1, :SR, krylovdim=kdim);
    @assert infos.converged > 0;
    lE‚ÇÄ[j]     = infos.converged>=1 ? vals[1] + V_nucl_nucl(R_min + j*Œ¥R)  : NaN;
    # on r√©cup√®re l'√©nergie  propre du niveau fondamental sur la tranche √† R fix√©
end  


# CALCUL DU R‚ÇÄ ET DES RAIDEURS
E‚ÇÄ_at_R‚ÇÄ, ind_R‚ÇÄ = findmin(lE‚ÇÄ);       # trouver l'√©nergie de surface minimale
R‚ÇÄ               = ind_R‚ÇÄ*Œ¥R + R_min;  # d√©finir le param√®tre nucl√©aire minimisant l'√©nergie de surface
‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ = 1/(Œ¥R)^2 * dot(diff2_ctr_o8, view(lE‚ÇÄ, ind_R‚ÇÄ-4:ind_R‚ÇÄ+4));  # on calcule la d√©riv√©e seconde  √† l'ordre 8 par rapport √† y de E‚ÇÄ en R‚ÇÄ
K = ‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ; # constante de raideur sur l'axe (Oy) pour le hamiltonien non perturb√©

œµ = 1/sqrt(sqrt(‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ*M)); # param√®tre de redimensionnement
œµ¬≤ = œµ^2;
u‚ÇÄ = 0.0;


# RESCALING: principe
# chgt variable R = R‚ÇÄ + œµu
# --> u ‚àà [1/œµ*(R_min-R‚ÇÄ), 1/œµ*(R_max-R‚ÇÄ)]
# donc notre nouvelle variable √† partir d'ici devient u
# V(R) = V(R‚ÇÄ+œµu) = VÃÉ(u)
# Œ®(X) = Œ®~(u)   

# NOUVEAUX PARAMETRES SUR AXE avec rescaling
u_min = 1/œµ*(R_min-R‚ÇÄ);
u_max = 1/œµ*(R_max-R‚ÇÄ);
Œ¥u = (u_max-u_min)/(N-1);
Œ¥u¬≤ = Œ¥u^2;
us = Vector(u_min:Œ¥u:u_max);
U = us' .* ones(N); # √©quivalent de Y: sur la grille donne (r,u) ‚Ü¶ u
Œõu = -1/œµ^2/Œ¥u*LS; # laplacien sur l'axe (Ou)

# FONCTIONS POTENTIELS NON PERTURB√âS SUR AXES S√âPAR√âS  
V‚ÇÄr  = V[:,ind_R‚ÇÄ];                         # formule 3.19 rapport sans la constante .- E‚ÇÄ_at_R‚ÇÄ
V‚ÇÄR  = .5*‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ*(Rs.-R‚ÇÄ).^2;     # formule 3.19 rapport sans la constante .+ E‚ÇÄ_at_R‚ÇÄ
V‚ÇÄu  = .5*‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ*œµ^2*us.^2;       # traduction en u idem sans la constante
@assert V‚ÇÄu ‚âà V‚ÇÄR;


# OP√âRATEURS POTENTIELS NON PERTURB√âS SUR AXES S√âPAR√âS 
VÃÇ‚Å∞r  = sparse(Diagonal(V‚ÇÄr));
VÃÇ‚Å∞R  = sparse(Diagonal(V‚ÇÄR));
VÃÇ‚Å∞u  = sparse(Diagonal(V‚ÇÄu)); 

# HAMILTONIENS NON PERTURB√âS SUR AXES S√âPAR√âS
HÃÇ‚Å∞r  = Œõr + VÃÇ‚Å∞r;
HÃÇ‚Å∞R  = ŒõR + VÃÇ‚Å∞R;
HÃÇ‚Å∞u  = Œõu + VÃÇ‚Å∞u;

# FONCTION POTENTIEL NON PERTURB√â SUR GRILLE
# formule 3.19: somme des deux premi√®res lignes en potentiel:
V‚ÇÄ     = @. V[:,ind_R‚ÇÄ]*ones(N)'    + .5*‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ*    (Rg.-R‚ÇÄ).^2; # (r,R) ‚Ü¶ V(r,R‚ÇÄ) + 1/2*(‚àÇ¬≤E‚ÇÄ/‚àÇR¬≤)(R‚ÇÄ)(R-R‚ÇÄ)¬≤
V‚ÇÄ_res = @. V[:,ind_R‚ÇÄ]*ones(N)'    + .5*‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ*œµ^2*  U     .^2; # (r,u) ‚Ü¶ V(r,u‚ÇÄ) + 1/2*(‚àÇ¬≤E‚ÇÄ/‚àÇu¬≤)(u‚ÇÄ)(u-u‚ÇÄ)¬≤
@assert V‚ÇÄ ‚âà V‚ÇÄ_res


# OP√âRATEUR POTENTIEL NON PERTURB√â SUR GRILLE
VÃÇ‚Å∞_res = Diagonal(reshape(V‚ÇÄ_res,  N¬≤));
VÃÇ‚Å∞     = Diagonal(reshape(V‚ÇÄ,      N¬≤));
# VÃÇ‚ÇÄ ‚âà VÃÇ‚ÇÄ_res = true

# CONSTRUCTION DU POTENTIEL ET DU HAMILTONIEN ORIGINAUX SUR GRILLE
VÃÉ = @. V_nucl_el(rg, 1/œµ*(Rg.-R‚ÇÄ)) .+ V_nucl_nucl(1/œµ*(Rg.-R‚ÇÄ)); # (r,u) ‚Ü¶ V(r,u)
HÃÇ_res, Œõ_res, VÃÇ_res = hamiltonian_2D_rescaled(Œ¥u, Œ¥u¬≤, Œ¥r, Œ¥r¬≤, N, N¬≤, V‚ÇÄ_res, m, œµ^2, K)


# CONSTRUCTION DU POTENTIEL ET DU HAMILTONIEN NON PERTURB√âS SUR GRILLE
HÃÇ‚Å∞     = Œõ     + VÃÇ‚Å∞;      # avec param√©trage R
HÃÇ‚Å∞_res = Œõ_res + VÃÇ‚Å∞_res;  # avec param√©trage u

# FONCTION PERTURBATION Vp (d√©veloppement de Taylor de la perturbation)
# ordre 1 en espace: (‚àÇV/‚àÇR)(r,R‚ÇÄ)√ó(R-R‚ÇÄ)
‚àÇR_of_V_at_rR‚ÇÄ   = 1/Œ¥R*  V[:,ind_R‚ÇÄ-4:ind_R‚ÇÄ+4] * diff1_ctr_o8';       # vecteur contenant ‚àÇV/‚àÇR(r,R‚ÇÄ) pour tous les r
V‚ÇÅ               = ‚àÇR_of_V_at_rR‚ÇÄ*ones(N)' .* (Rg.-R‚ÇÄ);                 # on le transforme en grille

# ordre 2 en espace: 1/2√ó[(‚àÇ¬≤V/‚àÇR¬≤)(r,R‚ÇÄ) - d¬≤E‚ÇÄ/dR¬≤(R‚ÇÄ)](R-R‚ÇÄ)¬≤ 
‚àÇ¬≤RR_of_V_at_rR‚ÇÄ = 1/Œ¥R¬≤* V[:,ind_R‚ÇÄ-4:ind_R‚ÇÄ+4] * diff2_ctr_o8';       # vecteur contenant ‚àÇ¬≤V/‚àÇR¬≤(r,R‚ÇÄ) pour tous les r
V‚ÇÇ               = .5*(‚àÇ¬≤RR_of_V_at_rR‚ÇÄ*ones(N)' .- ‚àÇ¬≤RR_of_E‚ÇÄ_at_R‚ÇÄ) .* (Rg.-R‚ÇÄ).^2; # on le transforme en grille

# perturbation totale ordre 1 + ordre 2
Vp     = V‚ÇÅ + V‚ÇÇ;
# OP√âRATEUR PERTURBATION WÃÇ
WÃÇ     = Diagonal(reshape(Vp, N¬≤)); # œà‚ààL¬≤(‚Ñù^N¬≤) ‚Ü¶ Vp√óœà param√©tr√© en R


# OP√âRATEURS RESCALED cf doc √âric
V_res = @. V_nucl_el(rg, R‚ÇÄ.+œµ*us) + V_nucl_nucl(R‚ÇÄ.+œµ*us); # (r,u) ‚Ü¶ V(r,u)

ùî•,~,~  = hamiltonian_2D_rescaled(Œ¥u, Œ¥u¬≤, Œ¥r, Œ¥r¬≤, N, N¬≤, V_res, m, œµ¬≤, K) # ùî• : Œ®(r,u) ‚àà L¬≤(‚Ñù^N^2) ‚Ü¶ -1/2m √ó ‚àÇ¬≤/‚àÇr¬≤ -1/2M √ó ‚àÇ¬≤/‚àÇu¬≤ + V(r,u) le hamiltonien original complet
ùî•u =   K*œµ¬≤*(-1/2/Œ¥u¬≤*LS + 1/2*sparse(Diagonal(us.^2)));  # ùî•u : œà(u) ‚àà L¬≤(‚Ñù^N) ‚Ü¶ 1/2 √ó Kœµ¬≤(-‚àÇ¬≤/‚àÇu¬≤ + u¬≤)œà pour la solution-produit
# on ne met pas +.E‚ÇÄ_at_R‚ÇÄ comme dans l'√©quation d'Eric pour garder la propri√©t√© sparse et acc√©l√©rer Krylov
ùî•r = - 1/2*Œõr + VÃÇ‚Å∞r;  

                                  # ùî•u : œà(r) ‚àà L¬≤(‚Ñù^N) ‚Ü¶ 1/2 √ó    (-‚àÇ¬≤/‚àÇr¬≤ + R¬≤)œà pour la solution-produit
ùî¥  =  @. V  - (V_nucl_el(rg, R‚ÇÄ) + V_nucl_nucl(Rg)) - 1/2*œµ¬≤*K*U.^2; # œà(r,u) ‚àà L¬≤(‚Ñù^N¬≤) ‚Ü¶ perturbation √† FINIR
ùî¥‚ÇÅ = œµ*‚àÇR_of_V_at_rR‚ÇÄ.*U;
ùî¥‚ÇÇ = .5*œµ^2*(‚àÇ¬≤RR_of_V_at_rR‚ÇÄ*ones(N)' .- K) .*U.^2;
Vp_res = ùî¥‚ÇÅ + ùî¥‚ÇÇ;
# @assert Vp_res ‚âà Vp;
WÃÇ_res  = Diagonal(reshape(Vp_res, N¬≤));
# œà‚ààL¬≤(‚Ñù^N¬≤) ‚Ü¶ Vp√óœà param√©tr√© en R

```

### 6) Calcul de la solution-produit s√©parable du hamiltonien non perturb√© $H_{HBO}$

```julia
nxc = 0; nyc = 0;
Œ®HBO_v = lœï‚Å∞x[nxc+1] * lœï‚Å∞y[nyc+1]'; # norm√© car les deux le sont
Œ®HBO = reshape(Œ®HBO_v,N*N)
EHBO = lE‚Å∞x[nxc+1] + lE‚Å∞y[nyc+1];
densit√© = N^2/(R_max-R_min)/(r_max-r_min);
heatmap(rs, Rs, densit√©*Œ®HBO_v'.^2,
    color=:greys,
    aspect_ratio=:auto,
    xlim=[r_min,r_max], ylim=[R_min, R_max],
    xlabel="coordonn√©e √©lectronique r",
    ylabel="coordonn√©e nucl√©aire R",
    size=(400,300),
    titlefontsize=10,
    title = "Mode (nx,ny)=($nxc,$nyc)\nŒ®HBO = œï(x)œà(y)\nEHBO = $(round(EHBO,digits=4))")
```

```julia
using IterativeSolvers
using LinearAlgebra
using CUDA
CUDA.reclaim()

Œ®‚ÇÄ = cu(convert(Vector{Float32}, Œ®HBO));
W = cu(convert(Matrix{Float32}, WÃÇ));
H‚ÇÄ = cu(convert(SparseMatrixCSC{Float32}, HÃÇ‚Å∞));



Œ†_par  = Œ®‚ÇÄ*Œ®‚ÇÄ';
Œ†_ort  = Diagonal(CUDA.ones(N*N)) - Œ†_par;
Qmax   = 5;

P_ort  = H‚ÇÄ*Œ†_ort; # produit sparse*dense fonctionne mais pas dans l'autre sens
P_ort  = Œ†_ort*P_ort - EHBO*Œ†_ort; # Œ†‚üÇ(H‚ÇÄ-E‚ÇÄ)Œ†‚üÇ
P_par  = H‚ÇÄ*Œ†_par;
P_par  = Œ†_par*P_par - EHBO*Œ†_par; # Œ†‚à•(H‚ÇÄ-E‚ÇÄ)Œ†‚à•

llŒ®    = zeros(Float32, N*N, Qmax);
llE    = zeros(Float32, Qmax);

R_ort  = -Œ†_ort*W; 
b      = R_ort*Œ®‚ÇÄ;
b_acc  = zeros(Float32, N*N);
Œ®_acc1 = CUDA.zeros(N*N);
Œ®_acc2 = CUDA.zeros(N*N);

# GC ordre 1 et sauvegarde
Œ®_acc1[:] = cg(P_ort, b);
Œ®_acc2[:] = Œ®_acc1;
llŒ®[:,1]  = Array(Œ®_acc1);
llE[1]    = Œ®‚ÇÄ'*W*Œ®‚ÇÄ;

R_par = -P_par*W + llE[1]*P_par;
R_ort +=           llE[1]*P_ort;



for q ‚àà 2:Qmax
    for i ‚àà 1:(q-1)
        llE[q] = llE[q] -1/2*dot(llŒ®[i], llŒ®[q-i]);
    end
    
    fill!(b_acc, 0);
    fill!(b,     0);
    for i ‚àà 1:(q-2)
        b_acc[:] = @views b_acc + llE[q-i]*llŒ®[:,i];
    end
    b[:] = cu(b_acc);
    b[:] = b + llE[q]*Œ®‚ÇÄ;
    Œ®_acc1[:] = cg(P_ort, Œ†_ort*b + R_ort*Œ®_acc1);
    Œ®_acc2[:] = cg(P_ort, Œ†_par*b + R_par*Œ®_acc2) + Œ®_acc1;
    llŒ®[:,q] = Array(Œ®_acc2);
end

Œ®_sol = copy(Œ®HBO);
for i in 1:Qmax
    Œ®_sol[:] = @views Œ®_sol + Œª^i*llŒ®[:,i]
end
```
